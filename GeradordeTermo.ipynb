```python
# @title
# Importe a biblioteca ipywidgets
!pip install python-docx

import ipywidgets as widgets
from docx import Document

def gerar_termo_e_salvar(nome, matricula, modelo, codigo_interno, numero_serie):
   document = Document("/content/modelo.docx")  # Substitua pelo caminho do seu modelo Word
   # Substitua os marcadores no modelo Word pelos valores fornecidos pelo usuário
   for p in document.paragraphs:
       if "<NOME>" in p.text:
           p.text = p.text.replace("<NOME>", nome)
       if "<MATRICULA>" in p.text:
           p.text = p.text.replace("<MATRICULA>", matricula)
       if "<MODELO>" in p.text:
           p.text = p.text.replace("<MODELO>", modelo)
       if "<CODIGO_INTERNO>" in p.text:
           p.text = p.text.replace("<CODIGO_INTERNO>", codigo_interno)
       if "<NUMERO_SERIE>" in p.text:
           p.text = p.text.replace("<NUMERO_SERIE>", numero_serie)

   nome_arquivo = f"{nome}_{matricula}_{numero_serie}.docx"  # Nome do arquivo com o formato especificado
   document.save(nome_arquivo)

   # Exibir mensagem de sucesso
   print(f"Termo de responsabilidade gerado e salvo como {nome_arquivo}")

def on_button_click(button):
   nome = nome_input.value
   matricula = matricula_input.value
   modelo = modelo_input.value
   codigo_interno = codigo_interno_input.value
   numero_serie = numero_serie_input.value

   gerar_termo_e_salvar(nome, matricula, modelo, codigo_interno, numero_serie)

   # Limpar os campos de entrada
   nome_input.value = ""
   matricula_input.value = ""
   modelo_input.value = ""
   codigo_interno_input.value = ""
   numero_serie_input.value = ""

# Crie os widgets de entrada
nome_input = widgets.Text(
   description='Nome:',
   placeholder='Digite seu nome',
   style={'description_width': 'initial'})
matricula_input = widgets.Text(
   description='Matrícula:',
   placeholder='Digite sua matrícula',
   style={'description_width': 'initial'})
modelo_input = widgets.Text(
   description='Modelo:',
   placeholder='Digite o modelo do equipamento',
   style={'description_width': 'initial'})
codigo_interno_input = widgets.Text(
   description='Código Interno:',
   placeholder='Digite o código interno do equipamento',
   style={'description_width': 'initial'})
numero_serie_input = widgets.Text(
   description='Número de Série:',
   placeholder='Digite o número de série do equipamento',
   style={'description_width': 'initial'})

# Crie o botão
button = widgets.Button(
   description='Gerar e Salvar Termo',
   disabled=False,
   button_style='', # 'success', 'info', 'warning', 'danger' or ''
   tooltip='Clique para gerar e salvar o termo de responsabilidade',
   icon='')

# Conecte o evento de clique do botão à função
button.on_click(on_button_click)

# Crie o layout
layout = widgets.VBox([nome_input, matricula_input, modelo_input, codigo_interno_input, numero_serie_input, button])

# Exiba o layout
display(layout)

```
